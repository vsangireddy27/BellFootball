```python
import pandas as pd

file_path1 = '/content/Serra vs Mitty.xlsx'
file_path2 = '/content/Serra vs SHC.xlsx'
file_path3 = '/content/Serra vs SI.xlsx'
file_path4 = '/content/Serra vs SJB.xlsx'
file_path5 = '/content/Serra vs Saint Francis.xlsx'
df1 = pd.read_excel(file_path1)
df2 = pd.read_excel(file_path2)
df3 = pd.read_excel(file_path3)
df4 = pd.read_excel(file_path4)
df5 = pd.read_excel(file_path5)

frames = [df1, df2, df3, df4, df5]
df = pd.concat(frames)

def calculate_coverage_stats(filtered_df):
    coverage_counts = filtered_df['COVERAGE'].value_counts().reset_index()
    coverage_counts.columns = ['COVERAGE', 'Frequency']
    total_plays = coverage_counts['Frequency'].sum()
    coverage_counts['Percentage'] = round((coverage_counts['Frequency'] / total_plays) * 100)
    coverage_counts['Ratio'] = coverage_counts['Frequency'].astype(str) + '/' + str(total_plays)
    return coverage_counts

def first_DN_10(df):
    filtered = df[(df['DN'] == 1) & (df['DIST'] == 10)]
    return calculate_coverage_stats(filtered)

def second_DN_10_plus(df):
    filtered = df[(df['DN'] == 2) & (df['DIST'] >= 10)]
    return calculate_coverage_stats(filtered)

def second_DN_5_10(df):
    filtered = df[(df['DN'] == 2) & (df['DIST'] >= 5) & (df['DIST'] < 10)]
    return calculate_coverage_stats(filtered)

def second_DN_0_5(df):
    filtered = df[(df['DN'] == 2) & (df['DIST'] < 5)]
    return calculate_coverage_stats(filtered)

def third_DN_10_plus(df):
    filtered = df[(df['DN'] == 3) & (df['DIST'] >= 10)]
    return calculate_coverage_stats(filtered)

def third_DN_5_10(df):
    filtered = df[(df['DN'] == 3) & (df['DIST'] >= 5) & (df['DIST'] < 10)]
    return calculate_coverage_stats(filtered)

def third_DN_0_5(df):
    filtered = df[(df['DN'] == 3) & (df['DIST'] < 5)]
    return calculate_coverage_stats(filtered)

def fourth_DN_5(df):
    filtered = df[(df['DN'] == 4) & (df['DIST'] < 5)]
    return calculate_coverage_stats(filtered)

def fourth_DN_plus5(df):
    filtered = df[(df['DN'] == 4) & (df['DIST'] >= 5)]
    return calculate_coverage_stats(filtered)

def analyze_coverage_frequencies(df):
    print("1st DN and 10 Yards:")
    print(first_DN_10(df))
    print("\n2nd DN and >= 10 Yards:")
    print(second_DN_10_plus(df))
    print("\n2nd DN and 5 - 10 Yards:")
    print(second_DN_5_10(df))
    print("\n2nd DN and 0 - 5 Yards:")
    print(second_DN_0_5(df))
    print("\n3rd DN and >= 10 Yards:")
    print(third_DN_10_plus(df))
    print("\n3rd DN and 5 - 10 Yards:")
    print(third_DN_5_10(df))
    print("\n3rd DN and 0 - 5 Yards:")
    print(third_DN_0_5(df))
    print("\n4th DN and <5 Yards:")
    print(fourth_DN_5(df))
    print("\n4th DN and >=5 Yards:")
    print(fourth_DN_plus5(df))

df_d_plays = df[df['ODK'] == 'D']
analyze_coverage_frequencies(df_d_plays)
"""
compiled_df = analyze_coverage_frequencies(df_d_plays)
output_path = '/content/Jesuit Coverage.xlsx'
compiled_df.to_excel(output_path, index=False)
"""
```
